```python
# Define input variables
temperature_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'temperature_my')
flavor_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'flavor_my')
portion_size_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'portion_size_my')

attentiveness_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'attentiveness_my')
friendliness_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'friendliness_my')
speed_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'speed_my')

# Define output variable
food_quality_my = ctrl.Consequent(np.linspace(0, 10, 11), 'food_quality_my')
service_quality_my = ctrl.Consequent(np.linspace(0, 10, 11), 'service_quality_my')

# Define membership functions
temperature_my.automf(3)
flavor_my.automf(3)
portion_size_my.automf(3)

food_quality_my.automf(3)

food_ctrl_my = ctrl.ControlSystem([
    ctrl.Rule(temperature_my['poor'] | flavor_my['poor'] | portion_size_my['poor'], food_quality_my['poor']),
    ctrl.Rule(temperature_my['average'] | flavor_my['average'] | portion_size_my['average'], food_quality_my['average']),
    ctrl.Rule(temperature_my['good'] | flavor_my['good'] | portion_size_my['good'], food_quality_my['good'])
])
```

```python
service_quality_my.automf(3)

attentiveness_my.automf(3)
friendliness_my.automf(3)
speed_my.automf(3)

service_ctrl_my = ctrl.ControlSystem([
    ctrl.Rule(attentiveness_my['poor'] | friendliness_my['poor'] | speed_my['poor'], service_quality_my['poor']),
    ctrl.Rule(attentiveness_my['average'] | friendliness_my['average'] | speed_my['average'], service_quality_my['average']),
    ctrl.Rule(attentiveness_my['good'] | friendliness_my['good'] | speed_my['good'], service_quality_my['good'])
])
```

```python
food_quality_for_tip_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'food_quality_for_tip_my')
service_quality_for_tip_my = ctrl.Antecedent(np.linspace(0, 10, 11), 'service_quality_for_tip_my')
tip_my = ctrl.Consequent(np.linspace(0, 25, 26), 'tip_my')
tip_my.automf(3)

food_quality_for_tip_my.automf(3)
service_quality_for_tip_my.automf(3)

tip_ctrl_my = ctrl.ControlSystem([
    ctrl.Rule(food_quality_for_tip_my['poor'] & service_quality_for_tip_my['poor'], tip_my['poor']),
    ctrl.Rule(food_quality_for_tip_my['average'] & service_quality_for_tip_my['average'], tip_my['average']),
    ctrl.Rule(food_quality_for_tip_my['good'] & service_quality_for_tip_my['good'], tip_my['good'])
])


# Create control system simulations
food_sim_my = ctrl.ControlSystemSimulation(food_ctrl_my)
service_sim_my = ctrl.ControlSystemSimulation(service_ctrl_my)
tip_sim_my = ctrl.ControlSystemSimulation(tip_ctrl_my)
```

```python
# User input for food ratings
print("On a scale of 1 to 10, rate the following for your food:")
food_sim_my.input['temperature_my'] = float(input("Food temperature: "))
food_sim_my.input['flavor_my'] = float(input("Food flavor: "))
food_sim_my.input['portion_size_my'] = float(input("Portion size: "))
food_sim_my.compute()

# User input for service ratings
print("On a scale of 1 to 10, rate the following for the service:")
service_sim_my.input['attentiveness_my'] = float(input("Attentiveness: "))
service_sim_my.input['friendliness_my'] = float(input("Friendliness: "))
service_sim_my.input['speed_my'] = float(input("Speed of service: "))
service_sim_my.compute()
```

```python
# Compute tip recommendation
food_quality_output_my = food_sim_my.output['food_quality_my']
service_quality_output_my = service_sim_my.output['service_quality_my']
tip_sim_my.input['food_quality_for_tip_my'] = food_quality_output_my
tip_sim_my.input['service_quality_for_tip_my'] = service_quality_output_my
tip_sim_my.compute()
```

```python
tip_output_my = tip_sim_my.output['tip_my']
print("tip amount:", tip_output_my)
```
